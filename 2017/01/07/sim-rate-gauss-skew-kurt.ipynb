```python
%reload_ext autoreload
%matplotlib notebook
%load_ext autoreload
%autoreload 2
%reload_ext autoreload

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn as sns
```

```python
nv = np.arange(30, 500, 10)
```

```python
nn = len(nv)
nmax = nv[-1]
```

```python


from __future__ import division


NRUN = 10000

m3_sqr = np.zeros((NRUN, nn))
m4_minus3_sqr = np.zeros((NRUN, nn))


for r in xrange(NRUN):
    Xall = np.random.normal(size=nmax)
    
    for i in xrange(nn):
        n = nv[i]
        
        X = Xall[:n]
        Xmean = np.mean(X)
        
        m3_sqr[r, i] = np.square(np.mean(np.power(X - Xmean, 3)))
        m4_minus3_sqr[r, i] = np.square(np.mean(np.power(X - Xmean, 4)) - 3)



```

```python

plt.figure()
plt.plot(nv, np.mean(m3_sqr, axis=0))



plt.figure()
plt.plot(np.log(nv), np.log(np.mean(m3_sqr, axis=0)))
```

```python
plt.figure()
plt.plot(nv, np.mean(m4_minus3_sqr, axis=0))

plt.figure()
plt.plot(np.log(nv), np.log(np.mean(m4_minus3_sqr, axis=0)))
```
